```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/user/Downloads/Employee.csv")
print(data)
```

```python
data['Gender'].value_counts().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('values')
plt.title('Gender Distribution')
plt.show()
```

```python
data["Age"].value_counts().sort_index().plot(kind="bar")
plt.xlabel("Ages")
plt.ylabel("Values")
plt.title("Ages")
plt.show()
```
